```python
import seaborn as sns   
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('titanic')

df.info()
```

```python
df.head()
```

```python
df.isna().mean()
```

```python
df['age'] = df['age'].fillna(df.groupby('pclass')['age'].transform('median'))

df.isna().mean()
```

```python
df = df.drop(columns=['deck', 'who'])
df = df.dropna()

df.isna().mean()
```

```python
df.head()
```

```python
df.value_counts('embarked')
```

```python
df.memory_usage(deep=True)
```

```python
for column in ['survived', 'pclass', 'sibsp', 'parch']:
    df[column] = df[column].astype(np.int8)

df.memory_usage(deep=True)
```

```python
for column in ['sex', 'embarked', 'embark_town', 'alive']:
    df[column] = df[column].astype("category")

df.memory_usage(deep=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['sex', 'embarked', 'embark_town', 'alive'], drop_first=True)

df.head()
```

```python
mappings = {"First": 3, "Second": 2, "Third": 1}
df['class'] = df['class'].map(mappings)

df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1:]

X.head()
```

```python
y.head()
```

```python
X_test = X.iloc[800:]
X_train = X.iloc[:800]
y_test = y.iloc[800:]
y_train = y.iloc[:800]
```

```python
import torch
import torch.nn as nn   
from torch.utils.data import DataLoader, TensorDataset
import torch.optim as optim

for columns in X_train.columns:
    X_train[column] = X_train[column].astype(np.float32)

for c in y_train.columns:
    y_train[c] = y_train[c].astype(np.float32)

X = X_train.values
y = y_train.values

X = torch.tensor(X)
y = torch.tensor(y)

dataset = TensorDataset(X, y)
loader = DataLoader(dataset, batch_size=9, shuffle=True)

data_iter = iter(loader)
while True:
    try:
        features, labels = next(data_iter)
        print(f"Features: {features} \n\n Labels: {labels}")
    except StopIteration:
        print("Iterated through all batches")
        break
```
