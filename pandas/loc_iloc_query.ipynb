# DataFrame for Indexing and Selection

- .loc vs .iloc: Use .loc for label-based indexing and .iloc for integer-based. Always prefer these over "chained indexing" (df[col][row]).

- Boolean Indexing: Filtering data based on conditions. 
```python 
df.loc[(df['age'] > 25) & (df['status'] == 'Active'), 'target_column']
```
- .query(): A cleaner, more readable way to filter data using a string expression. Itâ€™s often faster for large datasets.

````python
import seaborn as sns
import pandas as pd
import numpy as np

df = sns.load_dataset('titanic')

df.head()
````

````python
df.loc[(df['sex'] == 'male') & (df['age'] > 25), 'embark_town']
````

````python
df.query("sex == 'male' and age > 25")['embark_town']
````

````python
# row, column
val = df.iloc[0:5, :]
val
````
