# Creating Series

Series are pandas' own dictionary depiction of a object that holds data. it has it's own indexes agaisnt which it stores a value.

```python
import pandas as pd
import numpy as np

s = pd.Series([100, 200, 300], name="Points", dtype=np.int64)

s
```

```python
s = pd.Series([100, 200, 300], index = ["a", "b", "c"])

s
```

```python
s = pd.Series({"name": "Abir", "age": 22, "city": "Delhi"})

s['city']
```

# DataFrame

DataFrame is pandas own way of showing 2D tabular data with rows and columns. We can do exploratory analysis, data indexinf, wrangling, grouping and seggregation, reshape and merging, time series analysis, etc using this framework

```python
df = pd.DataFrame([{"Name": "Abir", "Age": 23, "City": "New Delhi"},{"Name": "Maria", "Age": 19, "City": "Bengaluru"}])

df
```

```python
df = pd.DataFrame([["Abir", 23, "New Delhi"], ["Maria", 19, "Pune"]], columns=["Name", "Age", "City"])

df
```

```python
df = pd.DataFrame({
    "Name":["Abir", "Maria"],
    "Age": [23, 19],
    "City": ["Kolkata", "Bengaluru"]
})

df
```
