Even the most profitable model will degrade over time. Showing you know how to maintain a model separates senior candidates from juniors.

### Key Concepts:
- **Deployment Method**: How will the client use it? (e.g., Integrated into CRM, Standalone Dashboard, or real-time API).
- **Feedback Loop**: A mechanism to capture actual outcomes (what doctors actually did) to retrain the model on fresh data.

### Model Degradation (Drift):
1. **Data Drift**: The inputs change (e.g., the demographics or specialties of doctors in the database shift).
2. **Concept Drift**: The relationship between inputs and the target variable changes (e.g., a new competitor drug enters the market, and doctors' behavior alters regardless of their profile).

## 3. Cloud Deployment: AWS vs. Azure vs. GCP üèóÔ∏è

Regardless of the provider, the workflow is: 
**Save Model** ‚Üí **Containerize (Docker)** ‚Üí **Provision Compute** ‚Üí **Expose as API**.

| Cloud Provider | Service | Strengths | Best For... |
| :--- | :--- | :--- | :--- |
| **AWS** | SageMaker | Industrial strength, granular control. | Strong engineering teams wanting cost optimization. |
| **Azure** | Azure ML | Best UI, enterprise integration (Office 365/AD). | Traditional corporates, low-code tracking, OpenAI access. |
| **GCP** | Vertex AI | Developer-friendly, built-in Kubernetes/TensorFlow. | Data-native teams, heavy analytics, deep learning. |

### Standard Deployment Steps (Cross-Platform):
1. **Upload Artifacts**: Save model weights to a bucket (S3/GCS/Azure Blob).
2. **Register Model**: Version-control your model in a Registry.
3. **Configure Endpoint**: Set up hardware (CPU/GPU) and containers.
4. **Deploy & Scale**: Expose as a live HTTPS endpoint with auto-scaling.
